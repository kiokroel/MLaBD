```python
from sklearn.metrics import silhouette_score
from sklearn.metrics import davies_bouldin_score
from sklearn.metrics.cluster import rand_score
from sklearn.metrics import completeness_score

def metric (model,data,data_y):
    print("Rand_Score:", rand_score(model, data_y))
    print("Completeness Score:", completeness_score(model, data_y))
    print("Silhouette_score:", silhouette_score(data, model))
    print("Davies-Bouldin Score:", davies_bouldin_score(data, model))
```

## Свой KMeans

```python
import numpy

class KMeans:
    def __init__(self, n_clusters = 8, max_iter=300):

        self.n_clusters = n_clusters
        self.max_iter = max_iter

    def _distance(self, dots, centroid):
        return np.sqrt(np.sum((dots[:, np.newaxis] - centroid) ** 2, axis=2))
    
    def _new_centroids(self, X):
        distances = self._distance(X, self.centroids)
        clasters = np.argmin(distances, axis=1)
        return np.array([np.mean(X[clasters == i], axis=0) for i in range(self.n_clusters)])

    def fit(self, X):
        self.centroids = X[np.random.choice(X.shape[0], self.n_clusters)]
        for _ in range(self.max_iter):
            new_centroids = self._new_centroids(X)
            if np.all(self.centroids == new_centroids):
                break
            self.centroids = new_centroids
            
    def predict(self, X):
        distances = self._distance(X, self.centroids)
        return np.argmin(distances, axis=1)


kmeans = KMeans(n_clusters=5)
kmeans.fit(data1)
pred = kmeans.predict(data1)

plt.scatter(data1[:,0], data1[:,1], c= pred)
plt.scatter(kmeans.centroids[:, 0], kmeans.centroids[:, 1], c='r')
metric(pred,data1,data_y1)
```
