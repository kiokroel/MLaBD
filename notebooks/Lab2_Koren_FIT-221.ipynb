## Заполним столбцы, содержащие числовые значения:
    
- Original Release Date заполним в формате (Year of Release).01.01
- Streaming Release Date заполним самой ранней датой в столбце, а если фильм вышел позже то датой выхода
- Tomatometer Count заполним средним значением в году(если информация в году отсутствует,то средним значением за все время)
- Также поступим с Tomatometer Rating, Audience Rating, Audience Count, Tomatometer Top Critics Count, Tomatometer Fresh Critics Count, Tomatometer Rotten Critics Count 
       

```python
f = lambda x : str(x)+"-01-01"
movies_table['Original Release Date'] = movies_table['Original Release Date'].fillna(movies_table['Year of Release'].transform(f))
```

Для удобства поменяем тип данных в столбце Streaming Release Date и Original Release Date

```python
movies_table['Streaming Release Date'] = pd.to_datetime(movies_table['Streaming Release Date'],format = '%Y-%m-%d')
movies_table['Original Release Date'] = pd.to_datetime(movies_table['Original Release Date'],format = '%Y-%m-%d')
```

```python
date = movies_table['Streaming Release Date'].min()
g = lambda x: x if (x > date) else date
movies_table['Streaming Release Date'] = movies_table['Streaming Release Date'].fillna((movies_table['Original Release Date'].transform(g)))
```

```python
def average_by_year(ds, colum_year, colum_info):
    func = ds.groupby(colum_year)[colum_info].transform('mean')
    ds[colum_info] = ds[colum_info].fillna(func)
    ds[colum_info] = ds[colum_info].fillna(ds[colum_info].mean())
```

```python
average_by_year(movies_table, 'Oscar Year', 'Tomatometer Count')
movies_table['Tomatometer Count'] = movies_table['Tomatometer Count'].astype(int)
```

```python
average_by_year(movies_table, 'Oscar Year', 'Tomatometer Rating')
```

```python
average_by_year(movies_table, 'Oscar Year', 'Audience Count')
movies_table['Audience Count'] = movies_table['Audience Count'].astype(int)
```

```python
average_by_year(movies_table, 'Oscar Year', 'Audience Rating')
```

```python
average_by_year(movies_table, 'Oscar Year', 'Tomatometer Top Critics Count')
movies_table['Tomatometer Top Critics Count'] = movies_table['Tomatometer Top Critics Count'].astype(int)
```

```python
average_by_year(movies_table, 'Oscar Year', 'Tomatometer Fresh Critics Count')
movies_table['Tomatometer Fresh Critics Count'] = movies_table['Tomatometer Fresh Critics Count'].astype(int)
```

```python
average_by_year(movies_table, 'Oscar Year', 'Tomatometer Rotten Critics Count')
movies_table['Tomatometer Rotten Critics Count'] = movies_table['Tomatometer Rotten Critics Count'].astype(int)
```

```python
movies_table['IMDB Votes'] = movies_table['IMDB Votes'].transform(lambda x: str(x).replace(',',''))
movies_table['IMDB Votes'] = movies_table['IMDB Votes'].astype(int)
```

## Изучение и обработка выбросов

```python
outlier = movies_table[['IMDB Votes', 'Audience Count']]
outlier
```

```python

Q1 = outlier.quantile(0.25)
Q1
```

```python

Q3 = outlier.quantile(0.75)
Q3
```

```python

IQR = Q3-Q1

data_filtered = outlier[~((outlier < (Q1 - 1.5 * IQR)) |(outlier > (Q3 + 1.5 * IQR))).any(axis=1)]
data_filtered
```

```python
# Копирование индексов
index_list = list(data_filtered.index.values)
```

```python
movies_table_filtered = movies_table[movies_table.index.isin(index_list)]
movies_table_filtered.head()
```

```python
filtered_outlier = movies_table_filtered[['IMDB Votes', 'Audience Count']]

for col in filtered_outlier:
    movies_table_filtered.hist(col, bins=100)
```

**Заключение:**

- Определили нижний и верхний пределы выбросов, определив 25-й и 75-й процентили. Затем нашли межквартильный размах. Использовали эту информацию для фильтрации выбросов. 
- Судя по гистограме Votes, у многих фильмов значение голосов близко к нулю- 
С гисторамой Audience Count то же самое.

# Зависимость рейтинга от длительности

```python
IMDB = movies_table_filtered.pivot_table(index='IMDB Rating', values='Movie Time', aggfunc='mean')
Audience = movies_table_filtered.pivot_table(index='Audience Rating', values='Movie Time', aggfunc='mean')

IMDB = IMDB.reset_index()
Audience = Audience.reset_index()
```

```python
sns.lmplot(x="Movie Time", y="IMDB Rating", data=IMDB)
plt.title('IMDB Rating')
sns.lmplot(x="Movie Time", y="Audience Rating", data=Audience)
plt.title('Audience Rating')
```

```python
duration_IMDB = IMDB[['Movie Time','IMDB Rating']]
corr_IMDB = duration_IMDB.corr()
duration_Audience = Audience[['Movie Time','Audience Rating']]
corr_Audience = duration_Audience.corr()
```

```python
round(corr_IMDB, 3)
```

```python
round(corr_Audience, 3)
```

### Заключение:
 - У фильмов с большей длительностью оценки IMDB выше
 - Длительность фильма не влияет на оценку Audience
