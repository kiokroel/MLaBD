# LDA

```python
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.datasets import fetch_20newsgroups
n_samples = 2000
n_features = 1000
n_topics = 10
n_top_words = 20
```

```python
def print_top_words(model, feature_names, n_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print("Topic #%d:" % topic_idx)
        print(" ".join([feature_names[i]
                        for i in topic.argsort()[:-n_top_words - 1:-1]]))
```

# BoW

```python
import pandas as pd
data=pd.read_csv('../data/spam.csv', encoding='ISO-8859-1')[0:4]["v2"]
```

```python
from pprint import pprint

def create_bag_of_words(texts):
    bag_of_words_vectors = []
    unique_words = set(word for text in texts for word in text.split())
    word_dict = {word: i for i, word in enumerate(unique_words)}
    for text in texts:
        word_count = [0] * len(unique_words)
        for word in text.split():
            word_count[word_dict[word]] += 1
        bag_of_words_vectors.append(word_count)
    return bag_of_words_vectors, word_dict

vectors, words = create_bag_of_words(data)

for vector in vectors:
    print(vector)
pprint(words)
```

# TF-IDF

```python
from pprint import pprint
import math

def calculate_tf(word, text):
    words = text.split()
    word_count = words.count(word)
    total_words = len(words)
    return word_count / total_words

def calculate_idf(word, texts):
    text_count = len(texts)
    word_appearances = sum(1 for text in texts if word in text)
    if word_appearances == 0:
        return 0
    return math.log(text_count / word_appearances)

def calculate_tfidf(texts):
    unique_words = set(word for text in texts for word in text.split())
    words_dict = {word: index for index, word in enumerate(unique_words)}
    tfidf_vectors = []
    for text in texts:
        tfidf_vector = []
        for word in unique_words:
            tf = calculate_tf(word, text)
            idf = calculate_idf(word, texts)
            tfidf = tf * idf
            tfidf_vector.append(tfidf)
        tfidf_vectors.append(tfidf_vector)
    return tfidf_vectors, words_dict

vectors, words = calculate_tfidf(data)
for vector in vectors:
    print(vector)
pprint(words)
```
